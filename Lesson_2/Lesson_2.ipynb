__1.__ Представьте в виде __несократимой__ рациональной дроби:
### $$а)\,\,\, 0.(216);$$
### $$б)\,\,\, 1.0(01).$$

Решение:

a)

$$a = 0.(216)$$
$$1000a = 216.(216)$$
$$1000a = 216 + 0.(216)$$
$$1000a = 216 + a$$
$$999a = 216$$
$$a = 216/999 = 24/111 = 8/37$$
$$8/37 = 0.(216)$$

б)
$$b = 0.(01)$$
$$100b = 1.(01)$$
$$100b = 1 + 0.(01)$$
$$99b = 1$$
$$b = 1/99$$

$$c = 1.0(01)$$
$$10c = 10.(01)$$
$$10c = 10 + 0.(01)$$
$$10c = 10 + 1/99$$
$$10c = 991/99$$
$$c = 991/990$$
$$991/990 = 1.0(01)$$

__2*.__ Пусть $x =\frac{2}{21}$. Известно, что для некоторого натурального $k$ число $x$ записывается в $k$ - ичной системе счисления как $0.(13)_k = 0, 131313..._k$. Найдите $k$.

Решение:

$$x = 0.(13)_k$$
$$k^2x = 13.(13)_k$$
$$k^2x = 13_k + 0.(13)_k$$
$$k^2x = 13_k + x$$
$$k^2x - x = 13_k$$
$$(k^2 -1)x = 13_k$$
$$x = \frac{13_k}{k^2 -1}$$
$$x = \frac{1 ⋅ k^1 + 3 ⋅ k^0}{k^2 - 1}$$
$$x = \frac{k + 3}{k^2 - 1}$$
$$\frac{2}{21} = \frac{k +3}{k^2 - 1}$$
$$2k^2 - 2 = 21k +63$$
$$2k^2 -21k - 65 = 0$$
$$k_1 = -2,5$$
$$k_2 = 13$$
Ответ: $k =13$
